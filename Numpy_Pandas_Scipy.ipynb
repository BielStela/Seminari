```python
%matplotlib inline
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt

import seaborn as sns
sns.set_style('white')
```

```python
df = sns.load_dataset("iris")
```

```python
df.head()
```

```python
pd.plotting.scatter_matrix(df, figsize=(10,10), diagonal='kde');
```

```python
sns.pairplot(df, hue='species', diag_kind='kde', plot_kws=dict(s=15))
```

```python
g = sns.PairGrid(df, diag_sharey=False, hue='species')
g.map_lower(sns.kdeplot, cmap="Blues_d")
g.map_upper(plt.scatter, s= 15)
g.map_diag(sns.kdeplot, lw=2, shade=True)
```

```python
plt.figure(figsize=(10,8))
sns.violinplot(data=df, palette="Set3", bw=.2, cut=1, linewidth=1,)
```

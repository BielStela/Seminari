```python
%matplotlib inline
import numpy as np
import pandas as pd
import seaborn as sns
sns.set_style('white')
from sklearn import datasets
```

```python
iris = datasets.load_iris()
```

```python
iris
```

```python
iris_df = pd.DataFrame(iris.data, columns=iris.feature_names)
```

```python
pd.plotting.scatter_matrix(iris_df, figsize=(10,10), diagonal='kde');
```

```python
colMap={0:"C0",1:"C1",2:"C2"}
cols=list(map(lambda x:colMap.get(x),iris.target))

pd.plotting.scatter_matrix(iris_df, figsize=(10,10), c = cols);
```

```python
iris_df['class'] = iris.target
pd.plotting.radviz(iris_df,'class',colormap='plasma')
```
